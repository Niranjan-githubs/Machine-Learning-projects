```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


```

```python
data = pd.read_csv('/content/sonar data.csv', header = None)
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.groupby(60).mean()
```

```python
X = data.drop(columns =60 , axis =1)
```

```python
Y = data[60]
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y, random_state= 1)
```

```python
model = LogisticRegression()
```

```python
model_train = model.fit(X_train,Y_train)
```

```python
model_prediction = model.predict(X_train)
```

```python
acc = accuracy_score(model_prediction,Y_train)
```

```python
print(acc)
```

```python
model_prediction_tes = model.predict(X_test)
```

```python
accuracy = accuracy_score(model_prediction_tes,Y_test)
```

```python
print(accuracy)
```
